```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train=app_train[[]'AMT_CREDIT_BINNED','']
# 連續資料離散化
app_train['AMT_CREDIT_BINNED'] = pd.cut(app_train['AMT_CREDIT'], 
                                  bins = np.linspace(start=app_train['AMT_CREDIT'].min(),stop=app_train['AMT_CREDIT'].max(),num = 11)) #自 20 到 70 歲，切 11 個點 (得到 10 組)
print(app_train['AMT_CREDIT_BINNED'].value_counts())
app_train.head(5)
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
app_train=app_train[['HOUSETYPE_MODE','AMT_CREDIT']]


unique_house_type = np.sort(app_train['HOUSETYPE_MODE'].unique())

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[app_train.HOUSETYPE_MODE==unique_house_type[i], 'AMT_CREDIT'].hist()
    
    plt.title("HouseType=["+str(unique_house_type[i])+"]")
plt.show()    
```
